##### **06 Para saber más: como crear un buen prompt con chat gpt**

Una de las habilidades más importantes para la interacción con IAs generativas como ChatGPT es la capacidad de crear buenos prompts, guiando la herramienta de acuerdo con tu conocimiento sobre el negocio. La creación de un prompt eficaz en ChatGPT es fundamental para obtener resultados precisos y relevantes para la resolución de problemas.

En el contexto de un proyecto de análisis de datos, como el que estamos trabajando aquí, un prompt bien elaborado puede ser la clave para obtener insights valiosos. A continuación, se presentan algunas buenas prácticas y ejemplos, teniendo en cuenta un proyecto simple de Ciencia de Datos:

***Sé específico y directo:***
Explica directamente lo que deseas que ChatGPT ejecute. Esto ayuda a la IA a entender mejor tu necesidad, proporcionando respuestas más precisas.

*Ejemplo*: “Realiza un análisis exploratorio de datos en la base ‘zoop_ventas.csv' usando las bibliotecas de manipulación de datos de Python. Presenta un resumen estadístico de las ventas por región”.

***Proporciona contexto cuando sea necesario:***
Informa a ChatGPT del contexto del proyecto, incluyendo datos, objetivos y restricciones. Esto puede mejorar significativamente la calidad de las respuestas. Sin embargo, evita información innecesaria que pueda confundir a la IA.

*Ejemplo*: "Estoy trabajando en un proyecto de análisis de ventas de una tienda minorista. La base de datos contiene información de ventas diarias por categoría de producto y región en 2023. Ayúdame a calcular la suma de las ventas de 'Electrónicos' por trimestre".

***Especifica el formato de la respuesta:***
Indica el formato deseado para la respuesta, como gráficos, tablas o explicaciones detalladas.

*Ejemplo*: "Crea un gráfico de líneas que muestre la evolución de las ventas a lo largo del último año, segmentado por categoría de productos".

***Indica el nivel de detalle deseado:***
Deja claro el nivel de detalle o profundidad de la explicación que estás buscando. Esto ayuda a la IA a ajustar la respuesta a tu necesidad de comprensión.

*Ejemplo*: "¿Podrías proporcionarme un ejemplo comentado de cómo usar la función groupby() de pandas para agrupar datos por categoría y calcular la suma total de ventas para cada grupo?"

***Usa términos técnicos correctos:***
Pon atención a la terminología específica relacionada con tu área de interés. Esto asegura que la IA comprenda el dominio de tu pregunta y responda de forma adecuada.

*Ejemplo*: “¿Cómo implemento un modelo de regresión lineal en Python usando scikit-learn para predecir ventas futuras con base en gastos de publicidad?”

***Divide tareas complejas en etapas simples:***
Si el proyecto involucra varias etapas, divídelo en partes más pequeñas e instruye a ChatGPT paso a paso. Esto facilita que la IA te ayude de manera eficaz.

*Ejemplo:*
"¿Cómo puedo manejar valores faltantes en mi conjunto de datos usando pandas?"<br>

"¿Cuál es la mejor manera de visualizar la distribución de edades en mi conjunto de datos usando matplotlib?"<br>

“¿Podrías ajustar el visual de la distribución de edades colocando las etiquetas y labels con tamaño 12 y color #CCCCCC?”<br>

***Mantente abierto a interacciones:***
Esté siempre preparado(a) para realizar ajustes en tus prompts basados en las respuestas recibidas. La interacción puede ayudar a refinar las respuestas y obtener exactamente lo que necesitas.

*Ejemplo*: “Esta respuesta puede mejorar. ¿Qué tal si redondeamos los valores a dos decimales y mostramos en el gráfico solo los valores mayores a 1 millón (1e6)?”

Vamos a actuar como analistas de datos de un e-commerce. Recibimos tres bases de datos con la información de los clientes y ventas en 2023 y qué clientes están pre-registrados en Zoop Pay, respectivamente.

Vamos a comenzar uniendo las dos primeras bases de nuestro proyecto ("ventas" y "clientes") en un DataFrame llamado "df" utilizando como clave la columna "ID_compra". Además, vamos a organizar las columnas en la siguiente secuencia:

["ID_compra", "fecha", "hora", "categoría", "precio_unitario", "cantidad", "envio", "metodo_pago", "ID_cliente", "edad", "sexo_biologico", "ciudad", "estado", "region", "cashback", "nota"]

Construye un código para ejecutar esta etapa usando Python y la biblioteca Pandas.

Utilizando la respuesta de ChatGPT, obtenemos el siguiente código:

````python
import pandas as pd

# Ejemplo de cómo cargar los DataFrames, si es necesario:
# ventas = pd.read_csv('ruta_al_archivo_de_ventas.csv')
# clientes = pd.read_csv('ruta_al_archivo_de_clientes.csv')

# Uniendo las bases "ventas" y "clientes" usando la columna "ID_compra" como clave
df = pd.merge(ventas, clientes, on="ID_compra")

# Organizando las columnas en la secuencia especificada
columnas_ordenadas = ["ID_compra", "fecha", "horario", "categoria", "precio_unitario", "cantidad", "envio", "metodo_pago", "ID_cliente", "edad", "sexo_biologico", "ciudad", "estado", "region", "cashback", "nota"]

df = df[columnas_ordenadas]

# Mostrando las primeras filas del DataFrame resultante para verificación
df.head()
````

Nota que en este código generado por ChatGPT necesitamos ajustar la parte que lee los datos de las bases, pasando los enlaces como creados al inicio del cuaderno de Desafíos.

Ahora, vamos a la segunda parte del problema, concatenando este DataFrame creado con la base Zoop Pay y ajustando la columna de fecha a formato datetime. Por lo tanto, podemos utilizar el siguiente prompt para la resolución de estos procesos:

Ahora, vamos a unir el "df" con la base "zoop_pay", leída a partir de la URL "url_zoop_pay", generando un nuevo "df". Para unir las bases, utilizaremos la columna "ID_cliente". Además, ajustaremos la columna "fecha" al formato datetime, donde su formato en la base de datos es "YYYY-MM-DD".

Utilizando la respuesta de ChatGPT, obtenemos el siguiente código:

````python
import pandas as pd

# Sustituye url_zoop_pay por la URL real proporcionada para la base zoop_pay
zoop_pay = pd.read_csv(url_zoop_pay)

# Uniendo el DataFrame df con la base zoop_pay utilizando la columna "ID_cliente"
df = pd.merge(df, zoop_pay, on="ID_cliente")

# Ajustando la columna "fecha" al formato datetime
df['fecha'] = pd.to_datetime(df['fecha'], format='%Y-%m-%d')

# Mostrando las primeras filas del DataFrame resultante para verificación
df.head()
````

### **2da Pregunta:** ¿Cuál fue la facturación por categoria en 2023?

**Observación:** Para la facturación, consideraremos aquí el producto entre el precio unitario y la cantidad, sumados al costo de envío de cada producto.


> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de barras horizontales que muestre la facturación por categoría de Zoop.

En "df", tenemos las siguientes columnas:
* categoria: categorías de los productos
* precio_unitario: precio de una unidad del producto
* cantidad: cantidad vendida del producto
* envio: costo de envío total de la compra.

Deseamos crear una visualización que muestre la facturación de cada categoría, donde la facturación se calcula multiplicando el precio unitario por la cantidad y sumando el costo de envío.

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
# 1. Crear columna de facturación por fila
df['facturacion'] = (df['precio_unitario'] * df['cantidad']) + df['envio']

# 2. Agrupar por categoría y sumar la facturación
facturacion_por_categoria = df.groupby('categoria')['facturacion'].sum().reset_index()

# 3. Ordenar de mayor a menor facturación
facturacion_por_categoria = facturacion_por_categoria.sort_values(by='facturacion', ascending=True)

# 4. Visualización: gráfico de barras horizontales
plt.figure(figsize=(10, 6))
sns.barplot(data=facturacion_por_categoria, x='facturacion', y='categoria', palette='viridis', orient='h')

# 5. Ajustes de título y ejes
plt.title('Facturación por Categoría de Productos (Zoop) - 2023')
plt.xlabel('Facturación Total (MXN$)')
plt.ylabel('Categoría')
plt.tight_layout()

# 6. Mostrar gráfico
plt.show()
````

En "df", tenemos las siguientes columnas:
* fecha: con las fechas de las compras en formato datetime (`aaaa-mm-dd`)
* facturación: calculada como el producto del precio unitario y la cantidad, sumado al costo de envío de cada venta.

Deseamos crear una visualización que muestre las ventas por mes. Primero, agrupa los datos por mes y luego crea una nueva columna llamada "mes" que contenga el nombre de cada mes, traduciendo los meses del índice mediante el siguiente diccionario. Utiliza la columna "fecha" para agrupar los datos y construir el gráfico. La columna "mes" solo debe usarse para cambiar las etiquetas del eje x.

```
meses = {
    'January': 'Ene',
    'February': 'Feb',
    'March': 'Mar',
    'April': 'Abr',
    'May': 'May',
    'June': 'Jun',
    'July': 'Jul',
    'August': 'Ago',
    'September': 'Sep',
    'October': 'Oct',
    'November': 'Nov',
    'December': 'Dic'
```

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Diccionario de traducción de meses
meses = {
    'January': 'Ene', 'February': 'Feb', 'March': 'Mar', 'April': 'Abr',
    'May': 'May', 'June': 'Jun', 'July': 'Jul', 'August': 'Ago',
    'September': 'Sep', 'October': 'Oct', 'November': 'Nov', 'December': 'Dic'
}

# 2. Asegurar que 'fecha' es tipo datetime
df['fecha'] = pd.to_datetime(df['fecha'])

# 3. Agrupar por mes y sumar la facturación
df_mensual = df.resample('M', on='fecha')['facturacion'].sum().reset_index()

# 4. Crear columna con el nombre del mes traducido
df_mensual['mes'] = df_mensual['fecha'].dt.month_name().map(meses)

# 5. Crear gráfico de líneas
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_mensual, x='mes', y='facturacion', marker='o', sort=False)

plt.title('Facturación Total Mensual - Zoop (2023)')
plt.xlabel('Mes')
plt.ylabel('Facturación (MXN$)')
plt.tight_layout()
plt.show()
````

* facturacion: con los valores de las ventas
* metodo_pago: con el método de pago elegido en la compra

````python
# 1. Asegurar que la columna 'fecha' esté en formato datetime
df['fecha'] = pd.to_datetime(df['fecha'])

# 2. Crear columna de trimestre (por año para diferenciar si hay más de un año)
df['trimestre'] = df['fecha'].dt.to_period('Q').astype(str)

# 3. Agrupar por trimestre y método de pago, sumando la facturación
ventas_trimestre = df.groupby(['trimestre', 'metodo_pago'])['facturacion'].sum().unstack().fillna(0)

# 4. Crear gráfico de barras apiladas
ventas_trimestre.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')

# 5. Ajustes de presentación
plt.title('Facturación Trimestral por Método de Pago por Trimestre - Zoop (2023)')
plt.xlabel('Trimestre')
plt.ylabel('Facturación Total (MXN$)')
plt.legend(title='Método de Pago')
plt.tight_layout()
plt.show()
````

La elección entre un histograma de columnas (también conocido como histograma de barras) y un gráfico de densidad generalmente depende del tipo de datos que se analizan y del objetivo de la visualización.

<u>¿Por qué un histograma de columnas?</u>

El histograma de columnas se utiliza ampliamente para visualizar la distribución de frecuencias o conteos en un conjunto de datos. Este tipo de gráfico es particularmente eficaz para mostrar la distribución de variables discretas o categorías. En el caso de las evaluaciones del público, que son valores enteros de 0 a 10, un histograma de columnas permite visualizar claramente cuántas evaluaciones se recibieron para cada punto específico. Esto facilita la identificación de tendencias, como la concentración de evaluaciones en ciertas puntuaciones, y ayuda a comprender la distribución general de las evaluaciones.

<u>Diferencia entre un Histograma de Columnas y un Gráfico de Densidad</u>

*Histograma de Columnas*: Utiliza barras verticales para representar la frecuencia de cada categoría o intervalo de valores. La altura de cada barra indica el conteo o la proporción de observaciones en esa categoría.

*Gráfico de Densidad*: Utiliza una línea para trazar la distribución continua de los datos. Este tipo de gráfico es más adecuado para variables continuas, donde muestra la probabilidad de valores dentro de un continuo, proporcionando una idea de la densidad de los datos.

**¿Cuándo elegir cada uno?**

*Histograma de Columnas*: Es ideal para datos categóricos o discretos, cuando se desea visualizar la frecuencia o el conteo por categoría. Es útil cuando las categorías son pocas y están claramente definidas, como las calificaciones de 0 a 10.

*Gráfico de Densidad*: Es más adecuado para variables continuas, cuando se desea entender la distribución y densidad de los datos a lo largo de un intervalo. Es útil para visualizar la forma de la distribución de los datos, como una distribución normal, binomial, etc.

**Ejemplos de Utilización en E-commerce**

*Histograma de Columnas*: Imagina que deseas analizar las evaluaciones de los clientes para un producto específico en tu e-commerce. Las evaluaciones van de 0 a 5. Un histograma de columnas puede mostrar cuántos clientes dieron cada puntuación, permitiendo identificar rápidamente la satisfacción general de los clientes con el producto.

*Gráfico de Densidad*: Si estás analizando el tiempo que los clientes pasan en tu sitio de e-commerce antes de realizar una compra, los datos serían más continuos (por ejemplo, minutos). Un gráfico de densidad podría mostrar la distribución del tiempo de navegación, ayudando a identificar si la mayoría de los clientes compra rápidamente o pasa mucho tiempo navegando antes de decidir.

<u>**Gráfico de columnas y barras horizontales**</u><br>
Un gráfico de columnas es ideal para la representación de valores numéricos de una o más categorías de datos, sean discretas o de intervalos definidos. Facilita una comparación rápida y directa entre diferentes categorías, siendo uno de los gráficos más utilizados en informes y presentaciones. El ejemplo a continuación demuestra este comportamiento:

````python
import matplotlib.pyplot as plt
categorias = ['A', 'B', 'C', 'D']
valores = [228, 118, 238, 110]
plt.figure(figsize=(8, 5))
bars = plt.bar(categorias, valores, color='skyblue')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 3, yval, ha='center', va='bottom')
plt.title('Cantidad Total de Productos Vendidos')
plt.tight_layout()
plt.show()
````

<u>**Gráfico de barras horizontales**</u><br> Es ideal cuando queremos presentar las categorías a la izquierda de forma horizontal, siendo la forma más fluida de leer textos que puedan ser extensos. El enfoque de este tipo de gráfico está primero en la categoría de los datos y, en segundo lugar, en los valores numéricos.

````python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter, MultipleLocator
marcas = ['Marca A', 'Marca B', 'Marca C']
valores = [1000000, 1300000, 850000]
plt.figure(figsize=(8, 5))
plt.barh(marcas, valores, color=['#4caf50', '#2196f3', '#ff9800'])
plt.gca().xaxis.set_major_formatter(FuncFormatter(lambda x, _: f'${int(x):,}'.replace(',', '.')))
plt.gca().xaxis.set_major_locator(MultipleLocator(500000))
plt.xlim(0, 1600000)
plt.xlabel('')
plt.ylabel('')
plt.title('FACTURACIÓN DE LAS MARCAS CON MAYOR PARTICIPACIÓN EN EL MERCADO', fontweight='bold')
plt.tight_layout()
plt.show()
````

<u>**Gráfico de columnas o barras agrupadas**</u><br>
Un gráfico de columnas o barras agrupadas es ideal cuando queremos comparar diferentes categorías respecto a una variable, mostrando cómo los valores se comparan dentro de cada categoría (en el ejemplo, los trimestres) en columnas separadas (en el ejemplo, productos E, D y F) y entre los diferentes grupos, facilitando la visualización de patrones y diferencias entre ellos.

````python
import pandas as pd
import matplotlib.pyplot as plt
datos = {
    '1er Trim': [210, 650, 760],
    '2do Trim': [900, 450, 380],
    '3er Trim': [395, 520, 205],
    '4to Trim': [380, 230, 625]
}
grupos = ['D', 'E', 'F']
df = pd.DataFrame(datos, index=grupos)
df_plot = df.T
ax = df_plot.plot(kind='bar', figsize=(10, 6), width=0.75)
plt.title('CANTIDAD DE PRODUCTOS VENDIDA POR TRIMESTRE', fontweight='bold')
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=0)
plt.legend(title='')
plt.tight_layout()
plt.show()
````

<u>**Gráfico de columnas o barras apiladas**</u><br>  Cuando queremos añadir a la comparación entre periodos o categorías, la comparación dentro de cada una. Este tipo de gráfico permite observar la composición de cada columna según la variable deseada (por ejemplo, las preferencias de consumo de café o té).

````python
import matplotlib.pyplot as plt
import numpy as np
categorias = ['X', 'Y', 'Z']
cafe = np.array([2000, 1000, 1250])
te = np.array([500, 1500, 1250])
bar_width = 0.5
fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(categorias, cafe, width=bar_width, label='Café', color='royalblue')
ax.bar(categorias, te, width=bar_width, bottom=cafe, label='Té', color='lightskyblue')
ax.set_ylim(0, 3000)
ax.set_title('PREFERENCIA DE CONSUMO EN DISTINTAS CIUDADES', fontsize=14, fontweight='bold')
ax.set_xlabel('')
ax.set_ylabel('')
ax.set_xticks(np.arange(len(categorias)))
ax.set_xticklabels(categorias)
ax.yaxis.set_visible(True)
ax.legend()
plt.tight_layout()
plt.show()
````

<u>**Gráfico de línea**</u><br>
Los gráficos de línea son ideales para representar tendencias a lo largo del tiempo o una secuencia ordenada. Cuando se tiene una gran cantidad de periodos, vale la pena usar gráficos de líneas para representar series temporales.

````python
import matplotlib.pyplot as plt
import pandas as pd
datos = {
    'Mes': ['ene-13', 'feb-23', 'mar-23', 'abr-23', 'may-23', 'jun-23',
            'jul-23', 'ago-23', 'sept-23', 'oct-23', 'nov-23', 'dic-23'],
    'Venta $': [1000, 800, 580, 700, 980, 600, 2000, 2400, 1950, 3000, 3500, 3100]
}
df = pd.DataFrame(datos)
plt.figure(figsize=(10, 5))
plt.plot(df['Mes'], df['Venta $'], marker='o', linestyle='-', color='teal')
plt.title('Evolución de Ventas Mensuales', fontsize=14, fontweight='bold')
plt.xlabel('Mes')
plt.ylabel('Venta ($)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

Los gráficos de líneas añaden una variable categórica a los análisis, permitiendo la comparación de series temporales en períodos de tiempo cortos o largos. En el ejemplo a continuación, observamos la facturación por región dividida por trimestres de un año cualquiera. Es posible comparar las facturaciones en cada región en diferentes momentos utilizando solo una visualización.

````python
import matplotlib.pyplot as plt
import pandas as pd

# Datos
datos = {
    'Trimestre': ['1° Tri', '2° Tri', '3° Tri', '4° Tri'],
    'Centro-Oeste': [14, 22, 12, 15],
    'Noreste': [25, 21, 23.5, 22],
    'Norte': [11, 24, 15, 20]
}

df = pd.DataFrame(datos)

# Crear gráfico
plt.figure(figsize=(10, 5))

plt.plot(df['Trimestre'], df['Centro-Oeste'], marker='o', label='Centro-Oeste', color='deepskyblue')
plt.plot(df['Trimestre'], df['Noreste'], marker='s', label='Noreste', color='darkblue')
plt.plot(df['Trimestre'], df['Norte'], marker='^', label='Norte', color='red')

# Título en negrita
plt.title('Facturación por Región por Trimestre', fontsize=14, fontweight='bold')

# Sin etiquetas de eje X, eje Y con etiqueta 'Millones' y desde 0
plt.xlabel('')
plt.ylabel('Millones')
plt.ylim(bottom=0)

# Estética
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
````

<u>**Gráfico de pastel**</u><br>
El gráfico de pastel se usa para representar proporciones de diferentes categorías o partes de un todo. Es ideal para composiciones estáticas, ya sea por valores numéricos o porcentajes. Hay que tener cuidado de no usar más de 5 categorías o categorías con proporciones muy similares o muy dispares

````python
import matplotlib.pyplot as plt

# Datos
marcas = ['A', 'B', 'C', 'D']
porcentajes = [20, 40, 25, 15]
colores = ['lightblue', 'deepskyblue', 'cornflowerblue', 'navy']

# Función personalizada para mostrar marca + porcentaje
def formato_autopct(pct, allvals):
    i = int(round(pct * len(allvals) / 100.0))
    label = marcas[i] if i < len(marcas) else ''
    return f'{label}\n{pct:.1f}%'

# Crear gráfico
fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(
    porcentajes,
    labels=None,  # Sin etiquetas externas
    colors=colores,
    autopct=lambda pct: formato_autopct(pct, porcentajes),
    startangle=90,
    wedgeprops={'edgecolor': 'white'}
)

# Título en negrita
ax.set_title('PARTICIPACIÓN DE LAS MARCAS DEL PRODUCTO B', fontsize=14, fontweight='bold')

# Asegura proporción circular
ax.axis('equal')

plt.tight_layout()
plt.show()
````

<u>**Gráfico de Cascada**</u><br> Los gráficos de cascada se usan para visualizar ajustes incrementales a un valor inicial, lo que los hace ideales para ver flujos o fluctuaciones de valores numéricos a lo largo del tiempo.

<u>**Histograma**</u><br>
Los histogramas se utilizan para observar la distribución de los datos en un intervalo dado. Estos intervalos, llamados "bins", muestran cuántas observaciones caen dentro de esa franja.

````python
import matplotlib.pyplot as plt
import pandas as pd

# Datos de la tabla proporcionada
data = {
    'EDAD': [10, 15, 20, 25, 30, 35, 40, 45, 50],
    'FRECUENCIA': [200, 210, 290, 380, 600, 710, 550, 310, 100]
}
df = pd.DataFrame(data)

# Para crear un histograma, necesitamos definir los "bins" (intervalos de edad).
# Dado que las edades están dadas en intervalos de 5 años, podemos usar esto como nuestros bins.
# Los bins deben abarcar desde el inicio de la primera barra hasta el final de la última.
# Por ejemplo, si la edad es 10, podemos considerarla el centro de un bin de [7.5, 12.5)
# O más simplemente, definir los bordes de los bins directamente a partir de los datos.
# Para edades discretas como estas, es más común usar un gráfico de barras,
# pero si realmente quieres un histograma que muestre la "distribución", necesitamos bins.

# Vamos a definir los bins de manera que cada "EDAD" esté en el centro de un bin.
# O, de forma más común para estas tablas de frecuencia, podemos usar un gráfico de barras.
# Si lo que buscas es cómo se distribuye la FRECUENCIA en cada EDAD específica,
# un gráfico de barras es más apropiado que un histograma en sentido estricto,
# ya que un histograma agrupa datos continuos en rangos.

# Sin embargo, si la EDAD se interpreta como el punto medio de un intervalo de clase,
# podemos inferir los límites de los bins.
# Por ejemplo, si 10 es el punto medio, el intervalo podría ser [7.5, 12.5).
# Vamos a crear bins explícitamente para que se parezca a un histograma tradicional
# donde cada "edad" es el inicio de un rango de 5 años, o el centro.

# Considerando que cada "EDAD" en tu tabla es el centro o un punto representativo
# de un intervalo, y los intervalos son de 5 en 5, podemos definir los bordes de los bins.
# Por ejemplo, para edad 10, el bin podría ser de 7.5 a 12.5.
# Para 15, de 12.5 a 17.5, etc.

bin_edges = np.arange(min(df['EDAD']) - 2.5, max(df['EDAD']) + 2.6, 5) # Generar los bordes de los bins

# Para que Matplotlib dibuje un histograma con frecuencias dadas para cada bin,
# en lugar de contar ocurrencias de datos brutos, se necesita un enfoque ligeramente diferente.
# Podríamos usar plt.bar si queremos barras que representen las frecuencias exactas en cada punto de edad.
# Pero si la instrucción es "histograma", implica la representación de bins contiguos.

# Una forma de simular esto para una tabla de frecuencias es crear datos sintéticos
# que tengan la distribución deseada y luego pasar esos datos a plt.hist().
# O, la forma más directa, usar plt.bar para cada 'edad' y hacer que parezca un histograma.
# Optaré por plt.bar, ya que representa directamente la tabla de frecuencias dada,
# y es lo que la mayoría de la gente espera al ver "frecuencia por edad" en un gráfico de barras con ejes contiguos.

fig, ax = plt.subplots(figsize=(10, 6))

# Dibujar las barras para cada edad
bars = ax.bar(df['EDAD'], df['FRECUENCIA'], width=4.5, color='skyblue', edgecolor='black') # Ancho de 4.5 para que las barras casi se toquen

# Configuración del título y etiquetas de los ejes
ax.set_title('Histograma de Frecuencia por Edad', fontsize=16, fontweight='bold')
ax.set_xlabel('Edad', fontsize=12, fontweight='bold')
ax.set_ylabel('Frecuencia', fontsize=12, fontweight='bold')

# Configurar los ticks del eje X para que coincidan con las edades
ax.set_xticks(df['EDAD'])
ax.tick_params(axis='x', labelsize=10, rotation=0) # Rotación 0 si no se superponen
ax.tick_params(axis='y', labelsize=10)

# Añadir etiquetas de valor encima de cada barra
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 10, round(yval), ha='center', va='bottom', fontsize=9, fontweight='bold')

# Mejorar la apariencia general
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False) # Mantener el eje Y visible con sus ticks
ax.spines['bottom'].set_visible(True) # Mantener el eje X visible

# Establecer límites del eje Y para un mejor espaciado
ax.set_ylim(0, max(df['FRECUENCIA']) * 1.15) # Un 15% de margen superior para las etiquetas

plt.tight_layout()
plt.show()
````

<u>**Gráfico de dispersión**</u><br>
Los gráficos de dispersión se utilizan para representar la relación o distribución entre dos variables numéricas. Cada punto en el gráfico representa una combinación de coordenadas, facilitando la identificación de patrones, tendencias y correlaciones importantes para el aprendizaje automático.

````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.ticker import FuncFormatter

# Datos de la tabla proporcionada
data = {
    'Tpo ent': [2, 3, 4, 5, 6, 6, 7, 7, 8, 8, 9, 10, 11, 12, 13, 14, 15],
    'Calificacion': [10, 9, 9, 9, 8, 7, 8, 7, 7, 6, None, 5, None, 4, 4, None, 4]
}
df = pd.DataFrame(data)

# Eliminar filas con valores NaN en 'Calificacion' para el gráfico de dispersión
df_cleaned = df.dropna(subset=['Calificacion']).copy()
df_cleaned['Calificacion'] = pd.to_numeric(df_cleaned['Calificacion'])

fig, ax = plt.subplots(figsize=(10, 6))

# Crear el gráfico de dispersión
ax.scatter(df_cleaned['Tpo ent'], df_cleaned['Calificacion'], s=100, alpha=0.7, color='purple', edgecolors='black')

# --- Agregar línea de tendencia punteada ---
z = np.polyfit(df_cleaned['Tpo ent'], df_cleaned['Calificacion'], 1)
p = np.poly1d(z)
ax.plot(df_cleaned['Tpo ent'], p(df_cleaned['Tpo ent']), "r--", linewidth=1.5, label='Línea de Tendencia')

# Configuración del título y etiquetas de los ejes
ax.set_title('TIEMPO DE ENTREGA v/s SATISFACCIÓN DEL CLIENTE', fontsize=16, fontweight='bold')
ax.set_xlabel('Tiempo de Entrega', fontsize=12, fontweight='bold')
ax.set_ylabel('Satisfacción del Cliente', fontsize=12, fontweight='bold')

# Configurar los ticks del eje X y Y para que sean más legibles
ax.set_xticks(range(min(df['Tpo ent']), max(df['Tpo ent']) + 1, 1))

# --- CORRECCIÓN AQUÍ: Convertir el valor a entero antes de pasarlo a range() ---
max_calificacion_int = int(df_cleaned['Calificacion'].max())
ax.set_yticks(range(0, max_calificacion_int + 1, 1))

ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)

# Añadir una cuadrícula para facilitar la lectura de los puntos
ax.grid(True, linestyle='--', alpha=0.6)

# Eliminar bordes para un aspecto más limpio
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Añadir leyenda para la línea de tendencia
ax.legend(loc='upper right')

plt.tight_layout()
plt.show()
````

1. <u>Uso de los colores</u>

El uso correcto de los colores en proyectos visuales es importante para garantizar legibilidad y uniformidad en los materiales, y para ello, deben observarse dos parámetros: contraste y consistencia.

**Contraste**:
El contraste es simplemente la diferencia entre dos cosas en un contexto comparativo. En la práctica, evaluamos el contraste relacionando colores, ya que no tiene sentido en un solo color. Contrastes más fuertes significan colores más diferentes entre sí, mientras que contrastes más débiles denotan colores más similares. Aquí es importante tener cuidado con esta proximidad porque los colores similares dificultan la comprensión de los gráficos.<br>
Usar un solo color como guía no es problemático, pero asegúrate de proporcionar suficiente contraste entre todos los tonos elegidos.

**Consistencia**:
En un proyecto con muchos gráficos debe existir un patrón en el uso de los colores. De lo contrario, los materiales no parecen formar parte del mismo conjunto y transmiten una sensación de falta de profesionalismo, sugiriendo que fueron hechos sin seguir un patrón bien definido.<br>
Los gráficos que siguen una misma orientación cromática, tienen un acabado más profesional.

2. <u>Tipografía</u>

Los gráficos brindan una disposición más visual a la información textual y ofrecen una presentación más clara de los datos que el texto corrido. Aun así, los gráficos estarán acompañados de algún texto, como títulos y leyendas, y deben ser legibles. Para ello, deben considerarse dos atributos relacionados con la tipografía:

**Tamaño**: Las palabras deben tener un tamaño mínimo para que se puedan leer sin dificultades. En trabajos impresos, usa al menos 10 pt y 20 pt para gráficos en presentaciones.

**Estilo**: Evita tipografías muy ornamentadas, cursivas o con características excesivamente acentuadas, ya que tienden a tener una legibilidad deficiente en dimensiones menores y en extensiones mayores de texto.

En caso de duda, da prioridad a las familias tipográficas más simples, como en el gráfico de la derecha.

3. <u>Jerarquía de la información</u>

¿Alguna vez has estado en una conversación en la que todos hablan al mismo tiempo y nadie se entiende? Lo mismo sucede en la comunicación visual si todo en la pantalla o en el papel quiere ser destacado al mismo tiempo: es necesario definir solo uno o dos elementos. De esta manera, colores y texto, especialmente, deben ser manipulados de manera que no saturen al lector con información.<br>

##### **Elementos de un gráfico en la biblioteca Matplotlib**
La biblioteca Matplotlib es una de las herramientas más populares y poderosas para la creación de gráficos y visualizaciones de datos en Python. Ofrece una amplia gama de funcionalidades y elementos que pueden ser personalizados para mejorar la legibilidad y la presentación de las visualizaciones de datos. Aquí están algunos de los elementos clave que puedes utilizar al crear gráficos con Matplotlib, acompañados de buenas prácticas y ejemplos orientados a un proyecto simple de Data Science.

1. xticks y yticks

Estas funciones se usan para personalizar los ticks de los ejes X e Y. Ajustar estos elementos puede mejorar la legibilidad de los gráficos, especialmente cuando se trabaja con datos que tienen escalas grandes o pequeñas, o cuando los valores están densamente distribuidos. Ejemplo: plt.xticks(range(0, 101, 10)) define los ticks del eje X para valores de 0 a 100, con intervalos de 10.

2. xlabel y ylabel

Son etiquetas para los ejes X e Y, respectivamente. Son esenciales para indicar lo que representa cada eje. Nombra tus ejes con etiquetas claras e informativas. Esto hace que el gráfico sea comprensible sin necesidad de consultar el contexto externo. Ejemplo: plt.xlabel('Tiempo (años)') y plt.ylabel('Población (millones)').

3. Título (title) y subtítulos

Agrega un título al gráfico, proporcionando una visión general inmediata de lo que el gráfico está mostrando. Los subtítulos pueden usarse para proporcionar información adicional o contextual. Usa un título descriptivo que pueda reducir la necesidad de etiquetas y considera los subtítulos para detalles adicionales, como la fuente de los datos o observaciones específicas. Ejemplo: plt.title('Crecimiento Poblacional a lo Largo del Tiempo').

4. Leyenda (legend)

La leyenda ayuda a identificar diferentes series o categorías dentro de un gráfico. Es crucial cuando un gráfico contiene múltiples líneas, barras u otros elementos visuales. Usa leyendas claras y coloca la leyenda de manera que no obstruya los datos. plt.legend(loc='best') elige automáticamente la mejor ubicación para la leyenda. Ejemplo: plt.legend(['Producto A', 'Producto B']).

5. Anotación (text y annotate)

La anotación permite agregar textos explicativos en puntos específicos del gráfico, lo que puede ser útil para resaltar características o puntos de datos importantes. text permite agregar texto en cualquier parte del gráfico, mientras que annotate se usa para añadir anotaciones con la opción de incluir una flecha que apunte a una parte específica del gráfico. Usa plt.annotate() para agregar anotaciones que aclaren puntos de datos inusuales o tendencias notables. Especifica la ubicación de la anotación y el punto de datos a ser anotado. Ejemplo: plt.annotate('Pico Máximo', xy=(2020, 500), xytext=(2021, 600), arrowprops=dict(facecolor='black')) destaca un pico en el gráfico.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Definición de colores para la identidad visual de Zoop
GRIS_1 = '#CCCCCC' # Color para el fondo de la figura y del gráfico
ROJO_1 = '#FF0000' # Color para las columnas del gráfico
GRIS_5 = '#333333' # Color para el título
AZUL_1 = '#0000FF' # Color para las etiquetas de los ejes y los totales
AZUL_3 = '#0066CC' # Nuevo color para el texto dentro del gráfico

# Preparación de los datos: Contar la cantidad de pagos por método
# Asegúrate de que 'metodo_pago' sea el nombre de la columna en tu DataFrame df
conteo_pagos = df['metodo_pago'].value_counts().reset_index()
conteo_pagos.columns = ['metodo_pago', 'Cantidad_Pagos'] # Renombramos las columnas para mayor claridad

# Opcional: Ordenar los métodos de pago por la cantidad (de mayor a menor)
conteo_pagos = conteo_pagos.sort_values(by='Cantidad_Pagos', ascending=False)

# Cálculo del porcentaje
# Sumamos las cantidades de los dos primeros métodos ordenados (los más utilizados)
if len(conteo_pagos) >= 2:
    suma_primeras_dos_columnas = conteo_pagos['Cantidad_Pagos'].iloc[0] + conteo_pagos['Cantidad_Pagos'].iloc[1]
else:
    suma_primeras_dos_columnas = 0

suma_todas_las_columnas = conteo_pagos['Cantidad_Pagos'].sum()

if suma_todas_las_columnas > 0:
    porcentaje = (suma_primeras_dos_columnas / suma_todas_las_columnas) * 100
else:
    porcentaje = 0

# Configuración del estilo de la figura y del gráfico
plt.rcParams['figure.facecolor'] = GRIS_1
plt.rcParams['axes.facecolor'] = GRIS_1

# Crear la figura y los ejes con el tamaño especificado
plt.figure(figsize=(10, 6))

# Usamos seaborn para crear el gráfico de barras.
ax = sns.barplot(x='metodo_pago', y='Cantidad_Pagos', data=conteo_pagos, palette=[ROJO_1])

# Personalización de Títulos y Etiquetas de Ejes
plt.title('Métodos de pago más utilizados en 2023 (Cantidad de Pagos)', fontsize=18, color=GRIS_5)

# Suprimir etiquetas y ticks del eje Y
ax.set_ylabel('')
ax.set_yticks([])

# Personalizar etiquetas y ticks del eje X
ax.set_xlabel('Método de Pago', fontsize=12, color=AZUL_1)
ax.tick_params(axis='x', labelsize=12, colors=AZUL_1)

# Ocultar Bordes del Gráfico (Spines)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color(AZUL_1)

# Mostrar la Cantidad de Pagos Encima de Cada Columna
for p in ax.patches:
    ax.annotate(f'{int(p.get_height()):,}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 10),
                textcoords='offset points',
                fontsize=12,
                color=AZUL_1)

# --- Añadir el texto con el porcentaje dentro del gráfico ---
# Obtenemos el ancho de una barra para calcular el desplazamiento
ancho_barra = ax.patches[0].get_width() if ax.patches else 0.8

# La posición x original era 1.5. Para correrla "el mismo espacio que una columna hacia la derecha", sumamos el ancho de una barra.
posicion_x_ajustada = 1.5 + ancho_barra

# Formatea el texto con el porcentaje en negrilla usando la sintaxis de Matplotlib (r'$\mathbf{...}$')
# 'r' antes de la cadena es para 'raw string', útil con backslashes.
# '\mathbf{}' es el comando para negrita en math text.
texto_a_mostrar = (
    r'$\mathbf{'+ f'{porcentaje:.2f}' + r'}$% de los clientes utilizan Tarjeta de Crédito o Transferencia ' +
    '\npara el pago. Sería interesante considerar asociaciones con ' +
    '\nbancos para la creación de un Zoop Pay con el fin de fidelizar ' +
    '\na nuestra clientela.'
)

ax.text(
    x=posicion_x_ajustada,
    y=2500, # Manteniendo la misma posición Y
    s=texto_a_mostrar,
    fontsize=10,
    color=AZUL_3,
    ha='center',
    va='center'
)

plt.tight_layout()
plt.show()
````

### **Gráfico 1 - Métodos de pago más utilizados en 2023:**

````python
def metodo_pago(df):
    import pandas as pd
    import matplotlib.pyplot as plt
    import seaborn as sns

    # Definición de colores para la identidad visual de Zoop
    GRIS_1 = '#CCCCCC' # Color para el fondo de la figura y del gráfico
    ROJO_1 = '#FF0000' # Color para las columnas del gráfico
    GRIS_5 = '#333333' # Color para el título
    AZUL_1 = '#0000FF' # Color para las etiquetas de los ejes y los totales
    AZUL_3 = '#0066CC' # Nuevo color para el texto dentro del gráfico

    # Preparación de los datos: Contar la cantidad de pagos por método
    # Asegúrate de que 'metodo_pago' sea el nombre de la columna en tu DataFrame df
    conteo_pagos = df['metodo_pago'].value_counts().reset_index()
    conteo_pagos.columns = ['metodo_pago', 'Cantidad_Pagos'] # Renombramos las columnas para mayor claridad

    # Opcional: Ordenar los métodos de pago por la cantidad (de mayor a menor)
    conteo_pagos = conteo_pagos.sort_values(by='Cantidad_Pagos', ascending=False)

    # Cálculo del porcentaje
    # Sumamos las cantidades de los dos primeros métodos ordenados (los más utilizados)
    if len(conteo_pagos) >= 2:
        suma_primeras_dos_columnas = conteo_pagos['Cantidad_Pagos'].iloc[0] + conteo_pagos['Cantidad_Pagos'].iloc[1]
    else:
        suma_primeras_dos_columnas = 0

    suma_todas_las_columnas = conteo_pagos['Cantidad_Pagos'].sum()

    if suma_todas_las_columnas > 0:
        porcentaje = (suma_primeras_dos_columnas / suma_todas_las_columnas) * 100
    else:
        porcentaje = 0

    # Configuración del estilo de la figura y del gráfico
    plt.rcParams['figure.facecolor'] = GRIS_1
    plt.rcParams['axes.facecolor'] = GRIS_1

    # Crear la figura y los ejes con el tamaño especificado
    plt.figure(figsize=(10, 6))

    # Usamos seaborn para crear el gráfico de barras.
    ax = sns.barplot(x='metodo_pago', y='Cantidad_Pagos', data=conteo_pagos, palette=[ROJO_1])

    # Personalización de Títulos y Etiquetas de Ejes
    plt.title('Métodos de pago más utilizados en 2023 (Cantidad de Pagos)', fontsize=18, color=GRIS_5)

    # Suprimir etiquetas y ticks del eje Y
    ax.set_ylabel('')
    ax.set_yticks([])

    # Personalizar etiquetas y ticks del eje X
    ax.set_xlabel('Método de Pago', fontsize=12, color=AZUL_1)
    ax.tick_params(axis='x', labelsize=12, colors=AZUL_1)

    # Ocultar Bordes del Gráfico (Spines)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_color(AZUL_1)

    # Mostrar la Cantidad de Pagos Encima de Cada Columna
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height()):,}',
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center',
                    xytext=(0, 10),
                    textcoords='offset points',
                    fontsize=12,
                    color=AZUL_1)

    # --- Añadir el texto con el porcentaje dentro del gráfico ---
    # Obtenemos el ancho de una barra para calcular el desplazamiento
    ancho_barra = ax.patches[0].get_width() if ax.patches else 0.8

    # La posición x original era 1.5. Para correrla "el mismo espacio que una columna hacia la derecha", sumamos el ancho de una barra.
    posicion_x_ajustada = 1.5 + ancho_barra

    # Formatea el texto con el porcentaje en negrilla usando la sintaxis de Matplotlib (r'$\mathbf{...}$')
    # 'r' antes de la cadena es para 'raw string', útil con backslashes.
    # '\mathbf{}' es el comando para negrita en math text.
    texto_a_mostrar = (
        r'$\mathbf{'+ f'{porcentaje:.2f}' + r'}$% de los clientes utilizan Tarjeta de Crédito o Transferencia ' +
        '\npara el pago. Sería interesante considerar asociaciones con ' +
        '\nbancos para la creación de un Zoop Pay con el fin de fidelizar ' +
        '\na nuestra clientela.'
    )

    ax.text(
        x=posicion_x_ajustada,
        y=2500, # Manteniendo la misma posición Y
        s=texto_a_mostrar,
        fontsize=10,
        color=AZUL_3,
        ha='center',
        va='center'
    )

    plt.tight_layout()

    return plt
````

### **Gráfico 2 - Total de ventas mensuales en 2023:**

````python
def ventas_mensuales(df):
  import pandas as pd
  import matplotlib.pyplot as plt
  import matplotlib.ticker as mticker

  # Definición de colores para la identidad visual de Zoop
  GRIS_1 = '#CCCCCC' # Color para el fondo de la figura y del gráfico
  GRIS_5 = '#333333' # Color para el título
  AZUL_1 = '#0000FF' # Color para la línea, etiquetas de ejes y ticks
  ROJO_1 = '#FF0000' # Color para el resaltado de puntos y la numeración

  # Definir el orden correcto de los meses y sus nombres abreviados en español
  meses_completo_a_abreviado = {
      'January': 'Ene', 'February': 'Feb', 'March': 'Mar',
      'April': 'Abr', 'May': 'May', 'June': 'Jun',
      'July': 'Jul', 'August': 'Ago', 'September': 'Sep',
      'October': 'Oct', 'November': 'Nov', 'December': 'Dic'
  }
  orden_meses_ingles = list(meses_completo_a_abreviado.keys())

  # --- Preparación de los datos ---
  # Asegúrate de que tu df ya tiene una columna 'fecha' en formato datetime
  # y una columna 'facturacion' con los valores de las ventas.

  df['mes'] = df['fecha'].dt.strftime('%B')
  ventas_mensuales = df.groupby('mes')['facturacion'].sum()
  ventas_mensuales = ventas_mensuales.reindex(orden_meses_ingles)

  # --- Identificar los 4 puntos de mayor volumen de ventas ---
  # Obtenemos los nombres de los meses que corresponden a los top 4 por volumen de ventas.
  top_4_meses_por_volumen = ventas_mensuales.nlargest(4).index.tolist()

  # Preparamos las coordenadas X y Y para los círculos de los top 4,
  # pero manteniendo el orden cronológico de los meses.
  top_4_x_coords = []
  top_4_ventas_valores = []
  meses_a_numerar = [] # Para almacenar los meses que son top 4 en orden cronológico

  for i, mes_ingles in enumerate(orden_meses_ingles):
      if mes_ingles in top_4_meses_por_volumen:
          top_4_x_coords.append(i)
          top_4_ventas_valores.append(ventas_mensuales.loc[mes_ingles])
          meses_a_numerar.append({'x': i, 'y': ventas_mensuales.loc[mes_ingles]})

  # --- Configuración de la Visualización ---

  # Configurar el color de fondo para toda la figura y el área del gráfico
  plt.rcParams['figure.facecolor'] = GRIS_1
  plt.rcParams['axes.facecolor'] = GRIS_1

  # Crear la figura y los ejes con el tamaño especificado
  plt.figure(figsize=(12, 6))
  ax = plt.gca() # Obtener los ejes para personalizar y añadir elementos

  # Usar Matplotlib para crear el gráfico de línea principal
  plt.plot(ventas_mensuales.index, ventas_mensuales.values,
          marker='o', color=AZUL_1, linewidth=2, zorder=2) # zorder para asegurar capas

  # --- Personalización de Títulos y Etiquetas ---

  # Título de la figura
  plt.title('Ventas totales mensuales en 2023', fontsize=18, color=GRIS_5)

  # Suprimir labels de los ejes X e Y
  plt.xlabel('')
  plt.ylabel('')

  # Personalizar etiquetas del eje X (xticks)
  plt.xticks(
      ticks=range(len(orden_meses_ingles)),
      labels=[meses_completo_a_abreviado[m] for m in orden_meses_ingles],
      fontsize=12,
      color=AZUL_1
  )
  plt.tick_params(axis='x', colors=AZUL_1)

  # Personalizar etiquetas del eje Y (yticks)
  def formato_moneda(x, pos):
      if x >= 1_000_000:
          return f"MXN$ {x/1_000_000:.2f} M"
      elif x >= 1_000:
          return f"MXN$ {x/1_000:.2f} K"
      else:
          return f"MXN$ {x:.2f}"

  formatter = mticker.FuncFormatter(formato_moneda)
  ax.yaxis.set_major_formatter(formatter)
  plt.tick_params(axis='y', labelsize=12, colors=AZUL_1)

  # --- Ocultar Bordes del Gráfico (Spines) ---
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)
  ax.spines['left'].set_color(AZUL_1)
  ax.spines['bottom'].set_color(AZUL_1)

  # --- Añadir círculos y numeración en los 4 puntos de mayor volumen ---
  marker_size_circle = 15 # Tamaño del círculo, ajusta si es necesario
  text_font_size = 12   # Tamaño de la fuente para la numeración, ajusta si es necesario

  # Plotear los círculos resaltados (sin línea de conexión)
  ax.plot(top_4_x_coords, top_4_ventas_valores,
          marker='o',
          markersize=marker_size_circle,
          linestyle='None',
          markeredgecolor=ROJO_1,
          markerfacecolor='None',
          markeredgewidth=2,
          zorder=4)

  # Añadir la numeración del 1 al 4, siguiendo el orden de los meses (izquierda a derecha)
  # El 'meses_a_numerar' ya está ordenado por la posición en el eje X
  for i, punto in enumerate(meses_a_numerar):
      numero = i + 1 # Genera 1, 2, 3, 4 en el orden de los puntos de izquierda a derecha
      x_pos = punto['x']
      y_pos = punto['y']

      # Ajustar la posición del texto para que esté a la izquierda del círculo.
      # El desplazamiento horizontal (ej. -0.3) dependerá del 'markersize' y el 'fontsize'.
      offset_x = -0.3 # Ajusta este valor para mover el número más a la izquierda/derecha
      offset_y = 0.0  # Ajusta este valor para mover el número más arriba/abajo

      ax.annotate(
          str(numero),
          xy=(x_pos, y_pos),
          xytext=(x_pos + offset_x, y_pos + offset_y),
          color=ROJO_1,
          fontsize=text_font_size,
          ha='right',
          va='center',
          zorder=5
      )

  # Ajustar el diseño para asegurar que todo encaje bien
  plt.tight_layout()
  # Mostrar el gráfico final

  return plt
````
